```python
%pylab inline
import kwant
```

```python
W_L = 10
L_S = 4
pot = 4.0

t = -2.5
E_F = 0 # Fermi energy
    
# Pauli matrices  
s_0 = np.array([[1 , 0], [0, 1]])
s_x = np.array([[0 , 1], [1, 0]])
s_y = np.array([[0, -1j], [1j, 0]])
s_z = np.array([[1, 0], [0, -1]])
```

```python
def make_system(W, L):
  
  # Lattice
  lat = kwant.lattice.general([(np.sqrt(3)*1/2, 1/2), (0, 1)], [(0, 0), (1/(2*np.sqrt(3)),1/2)], norbs = 2)
  
  def scattering_region(pos):
    x, y = pos
    return abs(x) < L/2 and abs(y) < W/2
  
  def lead_shape(pos):
    x, y = pos
    return abs(y) < W/2
  
  
  def hopping(sitei, sitej):
    xi, yi = sitei.pos
    xj, yj = sitej.pos
    phaseij = 0.0#2*np.pi*phi*(xi-xj)*((yi+yj)/2)
    return -t*s_0*np.exp(1j*phaseij)
  
  sys = kwant.Builder()
  sys[lat.shape(scattering_region, (0,0))] = -E_F*s_0
  sys[lat.neighbors()] = hopping
  
  # Attaching leads
  sym = kwant.TranslationalSymmetry(lat.vec((-2,1)))
  L_0 = kwant.Builder(sym, conservation_law = s_z)
  L_0[lat.shape(lead_shape, (0,0))] = -E_F*s_0
  L_0[lat.neighbors()] = hopping
  L_1 = L_0.reversed()
  
  sys.attach_lead(L_0)
  sys.attach_lead(L_1)
  return sys
```

```python
_sys = make_system(22, 14)
```

```python
fig = kwant.plot(_sys, dpi=196)
fig.savefig("Graphene_with_leads.png", dpi=196)
```

```python
def calculate_cond(sys, energy):
  # Finalize system
  sys = sys.finalized()
  
  # compute the scattering matrix at a given energy
  smatrix = kwant.smatrix(sys, energy)
  
  # compute the transmission probability from lead 0 to lead 1
  c = smatrix.transmission(1, 0)
  return c
```

```python
calculate_cond(_sys, 1.0)
```

```python
def calculate_transmissions(sys, es=None):
  
  if es is None:
    es = np.linspace(-0.4, 0.4, 100)
  
  sysf = sys.finalized()
  
  T00 = []
  for e in es:
    smatrix = kwant.smatrix(sysf, e)
    T00.append(smatrix.transmission(1, 0))
  
  return es, np.array(T00)
```

```python
es, ts = calculate_transmissions(_sys)
```

```python
plt.figure(figsize=(10, 6), dpi=112)
plt.plot(es, ts, '-+')
plt.xlabel("energy [t]")
plt.ylabel("conductance [e^2/h]")
plt.savefig("Graphene_conductance_w22_l14", dpi=112)
```

```python
es2, ts2 = calculate_transmissions(_sys, np.append(np.flip(-np.geomspace(0.001, 0.4, 50)), np.geomspace(0.001, 0.4, 50)))
```

```python
def conductanceForRatio(R, energy):
  
  W = 100
  L = np.floor(W/float(R))
  Rnew = float(W/L)
  
  syst = make_system(W, L)
  return Rnew, calculate_cond(syst, e)
```
