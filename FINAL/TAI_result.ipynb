```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
```

```python
Ws = np.loadtxt("Ws.txt")

e_1 = -20e-3
e_2 = 25e-3
e_3 = 9e-3
```

```python
p1g1 = np.loadtxt("params_1_G1_nsamples=4.txt")
p1g2 = np.loadtxt("params_1_G2_nsamples=4.txt")
p1g3 = np.loadtxt("params_1_G3_nsamples=4.txt")
```

```python
plt.figure(figsize=(8,6), dpi=196)
plt.errorbar(Ws*1e3, [g[0] for g in p1g1], [g[1] for g in p1g1], linestyle = '-', marker='^', capsize=2,
            label=r"$E_F={}$ meV".format(1e3*e_1), alpha=0.8)
plt.errorbar(Ws*1e3, [g[0] for g in p1g2], [g[1] for g in p1g2], linestyle = '-', marker='o', capsize=2,
            label=r"$E_F={}$ meV".format(1e3*e_2), alpha=0.8)
plt.errorbar(Ws*1e3, [g[0] for g in p1g3], [g[1] for g in p1g3], linestyle = '-', marker='s', capsize=2,
            label=r"$E_F={}$ meV".format(1e3*e_3), alpha=0.8)
plt.xscale('log')
plt.legend(loc='upper right');
plt.xlabel(r"$W$ $(\textnormal{meV})$", fontsize=16)
plt.ylabel(r"$G$ $(e^2/h)$", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.savefig("tai_results_params1_nsamples=4.png", dpi=196)
```
